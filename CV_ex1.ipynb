```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('box.jpeg', cv2.IMREAD_GRAYSCALE)
edges = cv2.Canny(image, 70, 150)
lines = cv2.HoughLinesP(edges, 1, np.pi / 180, threshold=10, minLineLength=15, maxLineGap=50)
if lines is not None:
    image_color = cv2.imread('box.jpeg')
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(image_color, (x1, y1), (x2, y2), (0, 0, 255), 2) 
    plt.imshow(cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB))
    plt.title('Hough Line Transform')
    plt.show()
else:
    print("No lines were detected.")
```

```python
image = cv2.imread('box.jpeg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 50, 150)
lines = cv2.HoughLinesP(edges, 1, np.pi / 180, threshold=10, minLineLength=25, maxLineGap=50)
gray = np.float32(gray)
dst = cv2.cornerHarris(gray, 2, 3, 0.04)
dst = cv2.dilate(dst, None)
image_with_corners = image.copy()
image_with_corners[dst > 0.01 * dst.max()] = [0, 0, 255]
if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(image_with_corners, (x1, y1), (x2, y2), (0, 0, 255), 2) 
plt.imshow(cv2.cvtColor(image_with_corners, cv2.COLOR_BGR2RGB))
plt.title('Hough Transform and Harris Corner Detection')
plt.show()
```
